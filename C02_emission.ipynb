```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: Load the dataset
# Replace 'co2_emissions.csv' with the path to your dataset
df = pd.read_csv('co2_emissions.csv')

# Step 4: Data Cleaning and Preparation
# Check for missing values
df.dropna(inplace=True)

# Convert Year to datetime if it's not already
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Step 5: Analyze the Data
# Calculate total CO2 emissions per year
total_emissions_per_year = df.groupby('Year')['CO2_Emissions'].sum().reset_index()

# Calculate CO2 emissions per country
emissions_per_country = df.groupby('Country')['CO2_Emissions'].sum().reset_index()

# Step 6: Visualization
# Plot total CO2 emissions over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='CO2_Emissions', data=total_emissions_per_year, marker='o')
plt.title('Total CO2 Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (in million metric tons)')
plt.grid(True)
plt.show()

# Plot CO2 emissions by country
plt.figure(figsize=(12, 8))
sns.barplot(x='CO2_Emissions', y='Country', data=emissions_per_country.sort_values(by='CO2_Emissions', ascending=False))
plt.title('CO2 Emissions by Country')
plt.xlabel('CO2 Emissions (in million metric tons)')
plt.ylabel('Country')
plt.show()
```
